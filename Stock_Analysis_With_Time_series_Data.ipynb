##Time Resampling


```python
from datetime import datetime
df_tesla.resample(rule='A').min()
#df_tesla.head()
```

```python
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

##Quarterly Start Frequency

```python
df_tesla.resample(rule='QS').max()['Open'].plot()
```

#Plotting Techniques


```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

#Rolling


```python
df_tesla['Open'].rolling(3).max().plot(figsize=(15,5))
```

```python
df_tesla['Open: 10 Days Rolling'] = df_tesla['Open'].rolling(10).mean()
df_tesla[["Open", "Open: 10 Days Rolling"]].plot(figsize=(15,5))
```

## Simple Moving Average
- Rolling


```python
df_tesla['Open: 10 Days Rolling']=df_tesla['Open'].rolling(window=10, min_periods=1).mean()
df_tesla['Open: 20 Days Rolling']=df_tesla['Open'].rolling(window=20, min_periods=1).mean()
df_tesla['Open: 30 Days Rolling']=df_tesla['Open'].rolling(window=30, min_periods=1).mean()
```

```python
df_tesla[['Open', 'Open: 10 Days Rolling']].plot(figsize=(15,5))
```

```python
df_tesla[['Open', 'Open: 10 Days Rolling','Open: 20 Days Rolling', 'Open: 30 Days Rolling']].plot(xlim=['2020-01-01', '2022-01-01'], figsize=(15,5))
```

##Cumulative Moving Average (CMA)
#Expanding

```python
df_tesla['Open'].expanding().mean().plot(figsize=(15, 5))
```

##EMA - Exponential Moving Average

1.   Smoothing factor -0.1
2.   List item



```python
df_tesla['EMA_0.1']= df_tesla['Open'].ewm(alpha=0.1, adjust='False').mean()
```

```python
df_tesla[['Open', 'EMA_0.1']].plot(xlim=['2020-01-01', '2022-01-01'], figsize=(15,5))
```

EWMA - Exponential Weighted Moving Average
Smoothing factor -0.1

```python
df_tesla['EMA from 5 Days Span']= df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open', 'EMA from 5 Days Span']].plot(xlim=['2020-01-01', '2022-01-01'], figsize=(15,5))
```

```python
df_tesla[['Open', 'EMA from 5 Days Span', 'EMA_0.1']].plot(xlim=['2020-01-01', '2022-01-01'], figsize=(15,5))
```
